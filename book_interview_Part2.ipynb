```python
import numpy as np
import matplotlib.pyplot as plt

# Function and gradient
def f(x):
    return (x - 3)**2 + 1

def grad_f(x):
    return 2 * (x - 3)

# Gradient descent path
eta = 0.25
x = 0.0
trajectory = [x]

for _ in range(8):
    x = x - eta * grad_f(x)
    trajectory.append(x)

# Plot curve
x_vals = np.linspace(-1, 6, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_vals, label="Loss function")

# Plot trajectory points and arrows
for i in range(len(trajectory) - 1):
    x0 = trajectory[i]
    x1 = trajectory[i + 1]
    y0 = f(x0)
    y1 = f(x1)

    plt.scatter(x0, y0, zorder=3)
    plt.annotate(
        "",
        xy=(x1, y1),
        xytext=(x0, y0),
        arrowprops=dict(arrowstyle="->", lw=1.5)
    )

plt.scatter(trajectory[-1], f(trajectory[-1]), zorder=3, label="Gradient descent steps")
plt.scatter(3, f(3), marker="x", s=100, label="Minimum")

plt.title("Gradient Descent Path")
plt.xlabel("Parameter")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig("//Figures/gradient_descent_path.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x, y):
    return x**2 + 2*y**2

# Gradient of the function
def grad(x, y):
    return np.array([2*x, 4*y])

# Grid for contour plot
x = np.linspace(-3, 3, 200)
y = np.linspace(-3, 3, 200)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Point where we evaluate gradient
point = np.array([1.5, 1.0])
gradient = grad(point[0], point[1])

# Normalize gradient for plotting
grad_dir = gradient / np.linalg.norm(gradient)

# Plot contours
plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=20)

# Plot the point
plt.scatter(point[0], point[1], color='black', zorder=3)

# Gradient (steepest ascent)
plt.quiver(
    point[0], point[1],
    grad_dir[0], grad_dir[1],
    angles='xy', scale_units='xy', scale=1,
    label="Gradient (steepest ascent)"
)

# Negative gradient (steepest descent)
plt.quiver(
    point[0], point[1],
    -grad_dir[0], -grad_dir[1],
    angles='xy', scale_units='xy', scale=1,
    label="Negative gradient (steepest descent)"
)

plt.title("Gradient Indicates Direction of Steepest Ascent")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.tight_layout()
plt.savefig("/Figures/gradient_steepest_ascent.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Loss surface
def f(x, y):
    return x**2 + 2*y**2

# True gradient
def grad(x, y):
    return np.array([2*x, 4*y])

# Batch GD trajectory
eta = 0.15
point = np.array([2.5, 2.0])
batch_path = [point.copy()]

for _ in range(20):
    point = point - eta * grad(point[0], point[1])
    batch_path.append(point.copy())

batch_path = np.array(batch_path)

# SGD-like noisy trajectory
point = np.array([2.5, 2.0])
sgd_path = [point.copy()]

for _ in range(20):
    noise = np.random.normal(scale=0.6, size=2)
    noisy_grad = grad(point[0], point[1]) + noise
    point = point - eta * noisy_grad
    sgd_path.append(point.copy())

sgd_path = np.array(sgd_path)

# Contours
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=20)

plt.plot(batch_path[:, 0], batch_path[:, 1], marker="o", label="Batch gradient descent")
plt.plot(sgd_path[:, 0], sgd_path[:, 1], marker="o", label="Stochastic gradient descent")

plt.scatter(0, 0, marker="x", s=100, label="Minimum")
plt.title("Batch Gradient Descent vs Stochastic Gradient Descent")
plt.xlabel("Parameter 1")
plt.ylabel("Parameter 2")
plt.legend()
plt.tight_layout()
plt.savefig/Figures/sgd_vs_batch.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)

def f(x, y):
    return x**2 + 2*y**2

def grad(x, y):
    return np.array([2*x, 4*y])

eta = 0.15
start = np.array([2.5, 2.0])

# Batch GD
point = start.copy()
batch_path = [point.copy()]
for _ in range(20):
    point = point - eta * grad(point[0], point[1])
    batch_path.append(point.copy())
batch_path = np.array(batch_path)

# Mini-batch GD (moderate noise)
point = start.copy()
mini_batch_path = [point.copy()]
for _ in range(20):
    noise = np.random.normal(scale=0.25, size=2)
    g = grad(point[0], point[1]) + noise
    point = point - eta * g
    mini_batch_path.append(point.copy())
mini_batch_path = np.array(mini_batch_path)

# SGD (larger noise)
point = start.copy()
sgd_path = [point.copy()]
for _ in range(20):
    noise = np.random.normal(scale=0.6, size=2)
    g = grad(point[0], point[1]) + noise
    point = point - eta * g
    sgd_path.append(point.copy())
sgd_path = np.array(sgd_path)

# Contours
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=20)

plt.plot(batch_path[:, 0], batch_path[:, 1], marker="o", label="Batch GD")
plt.plot(mini_batch_path[:, 0], mini_batch_path[:, 1], marker="o", label="Mini-batch GD")
plt.plot(sgd_path[:, 0], sgd_path[:, 1], marker="o", label="SGD")

plt.scatter(0, 0, marker="x", s=100, label="Minimum")
plt.title("Batch vs Mini-Batch vs Stochastic Gradient Descent")
plt.xlabel("Parameter 1")
plt.ylabel("Parameter 2")
plt.legend()
plt.tight_layout()
plt.savefig("/Figures/batch_vs_minibatch_vs_sgd.pdf")
plt.show()
```

## Momentum and Adaptive Optimisers

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + 10*y**2

def grad(x, y):
    return np.array([2*x, 20*y])

eta = 0.05
beta = 0.9

start = np.array([3.0, 2.0])

# Standard gradient descent
point = start.copy()
gd_path = [point.copy()]

for _ in range(30):
    g = grad(point[0], point[1])
    point = point - eta * g
    gd_path.append(point.copy())

gd_path = np.array(gd_path)

# Momentum
point = start.copy()
velocity = np.zeros(2)
mom_path = [point.copy()]

for _ in range(30):
    g = grad(point[0], point[1])
    velocity = beta*velocity + g
    point = point - eta*velocity
    mom_path.append(point.copy())

mom_path = np.array(mom_path)

# Contour plot
x = np.linspace(-3,3,200)
y = np.linspace(-3,3,200)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.figure(figsize=(8,6))
plt.contour(X,Y,Z,levels=20)

plt.plot(gd_path[:,0], gd_path[:,1], marker='o', label="Gradient Descent")
plt.plot(mom_path[:,0], mom_path[:,1], marker='o', label="Momentum")

plt.scatter(0,0,marker='x',s=100,label="Minimum")

plt.legend()
plt.title("Gradient Descent vs Momentum")
plt.xlabel("Parameter 1")
plt.ylabel("Parameter 2")

plt.tight_layout()
plt.savefig("/Figures/momentum_vs_gd.pdf")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

def f(x, y):
    return x**2 + 50*y**2

def grad(x, y):
    return np.array([2*x, 100*y])

eta = 0.03
beta = 0.9
eps = 1e-8

start = np.array([3.0, 2.0])

# Standard gradient descent
point = start.copy()
gd_path = [point.copy()]

for _ in range(40):
    g = grad(point[0], point[1])
    point = point - eta * g
    gd_path.append(point.copy())

gd_path = np.array(gd_path)

# RMSProp
point = start.copy()
s = np.zeros(2)
rms_path = [point.copy()]

for _ in range(40):
    g = grad(point[0], point[1])
    s = beta*s + (1-beta)*(g**2)
    point = point - eta * g / (np.sqrt(s) + eps)
    rms_path.append(point.copy())

rms_path = np.array(rms_path)

# Contour plot
x = np.linspace(-3.5, 3.5, 300)
y = np.linspace(-2.5, 2.5, 300)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(8,6))
plt.contour(X, Y, Z, levels=25)

plt.plot(gd_path[:,0], gd_path[:,1], marker='o', markersize=3, label="Gradient Descent")
plt.plot(rms_path[:,0], rms_path[:,1], marker='o', markersize=3, label="RMSProp")

plt.scatter(0,0,marker='x',s=120,label="Minimum")

plt.title("RMSProp Adapts Learning Rates Across Dimensions")
plt.xlabel("Parameter 1")
plt.ylabel("Parameter 2")
plt.legend()

plt.tight_layout()
plt.savefig(os.path.join(FIG_DIR, "rmsprop_adaptation.pdf"))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

def f(x, y):
    return x**2 + 50*y**2

def grad(x, y):
    return np.array([2*x, 100*y])

eta = 0.03
beta1 = 0.9
beta2 = 0.999
eps = 1e-8

start = np.array([3.0, 2.0])

# Gradient Descent
point = start.copy()
gd_path = [point.copy()]
for _ in range(40):
    g = grad(point[0], point[1])
    point = point - eta*g
    gd_path.append(point.copy())
gd_path = np.array(gd_path)

# Momentum
point = start.copy()
v = np.zeros(2)
mom_path = [point.copy()]
for _ in range(40):
    g = grad(point[0], point[1])
    v = beta1*v + g
    point = point - eta*v
    mom_path.append(point.copy())
mom_path = np.array(mom_path)

# Adam
point = start.copy()
m = np.zeros(2)
v = np.zeros(2)
adam_path = [point.copy()]

for t in range(1,41):
    g = grad(point[0], point[1])

    m = beta1*m + (1-beta1)*g
    v = beta2*v + (1-beta2)*(g**2)

    m_hat = m/(1-beta1**t)
    v_hat = v/(1-beta2**t)

    point = point - eta*m_hat/(np.sqrt(v_hat)+eps)
    adam_path.append(point.copy())

adam_path = np.array(adam_path)

# Contour plot
x = np.linspace(-3.5,3.5,300)
y = np.linspace(-2.5,2.5,300)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

plt.figure(figsize=(8,6))
plt.contour(X,Y,Z,levels=25)

plt.plot(gd_path[:,0],gd_path[:,1],marker='o',markersize=3,label="Gradient Descent")
plt.plot(mom_path[:,0],mom_path[:,1],marker='o',markersize=3,label="Momentum")
plt.plot(adam_path[:,0],adam_path[:,1],marker='o',markersize=3,label="Adam")

plt.scatter(0,0,marker='x',s=120,label="Minimum")

plt.legend()
plt.title("Optimisation Paths: GD vs Momentum vs Adam")
plt.xlabel("Parameter 1")
plt.ylabel("Parameter 2")

plt.tight_layout()
plt.savefig(os.path.join(FIG_DIR,"adam_vs_sgd.pdf"))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

# -----------------------------
# Rotated ill-conditioned quadratic
# -----------------------------
A = np.array([[3.0, 2.0],
              [2.0, 6.0]])

def f(theta):
    return 0.5 * theta.T @ A @ theta

def grad(theta):
    return A @ theta

# Starting point
start = np.array([4.0, 3.0])

# Hyperparameters
eta_gd = 0.18
eta_mom = 0.12
eta_adam = 0.25

beta = 0.9
beta1 = 0.9
beta2 = 0.999
eps = 1e-8
steps = 40

# -----------------------------
# Gradient Descent
# -----------------------------
theta = start.copy()
gd_path = [theta.copy()]

for _ in range(steps):
    g = grad(theta)
    theta = theta - eta_gd * g
    gd_path.append(theta.copy())

gd_path = np.array(gd_path)

# -----------------------------
# Momentum
# -----------------------------
theta = start.copy()
velocity = np.zeros_like(theta)
mom_path = [theta.copy()]

for _ in range(steps):
    g = grad(theta)
    velocity = beta * velocity + g
    theta = theta - eta_mom * velocity
    mom_path.append(theta.copy())

mom_path = np.array(mom_path)

# -----------------------------
# Adam
# -----------------------------
theta = start.copy()
m = np.zeros_like(theta)
v = np.zeros_like(theta)
adam_path = [theta.copy()]

for t in range(1, steps + 1):
    g = grad(theta)

    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * (g ** 2)

    m_hat = m / (1 - beta1 ** t)
    v_hat = v / (1 - beta2 ** t)

    theta = theta - eta_adam * m_hat / (np.sqrt(v_hat) + eps)
    adam_path.append(theta.copy())

adam_path = np.array(adam_path)

# -----------------------------
# Contour grid
# -----------------------------
x = np.linspace(-5, 5, 400)
y = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x, y)

Z = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        point = np.array([X[i, j], Y[i, j]])
        Z[i, j] = f(point)

# -----------------------------
# Plot
# -----------------------------
plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, levels=25)

plt.plot(gd_path[:, 0], gd_path[:, 1], marker='o', markersize=3, linewidth=1.5, label="Gradient Descent")
plt.plot(mom_path[:, 0], mom_path[:, 1], marker='s', markersize=3, linewidth=1.5, label="Momentum")
plt.plot(adam_path[:, 0], adam_path[:, 1], marker='^', markersize=3, linewidth=1.5, label="Adam")

plt.scatter(start[0], start[1], s=80, label="Start")
plt.scatter(0, 0, marker='x', s=120, label="Minimum")

plt.title("Optimisation Paths on a Rotated Loss Surface")
plt.xlabel("Parameter 1")
plt.ylabel("Parameter 2")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(FIG_DIR, "adam_vs_sgd.pdf"))
plt.show()
```

## Regularisation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import make_pipeline
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

np.random.seed(0)

X = np.linspace(-3,3,30)
y = np.sin(X) + np.random.normal(0,0.2,len(X))

X = X.reshape(-1,1)

# model without regularisation
model_no_reg = make_pipeline(
    PolynomialFeatures(12),
    LinearRegression()
)

# model with regularisation
model_reg = make_pipeline(
    PolynomialFeatures(12),
    Ridge(alpha=1.0)
)

model_no_reg.fit(X,y)
model_reg.fit(X,y)

X_test = np.linspace(-3,3,200).reshape(-1,1)

plt.figure(figsize=(8,5))
plt.scatter(X,y,label="data")

plt.plot(X_test,
         model_no_reg.predict(X_test),
         label="no regularisation")

plt.plot(X_test,
         model_reg.predict(X_test),
         label="with regularisation")

plt.legend()
plt.title("Effect of Regularisation")
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()

plt.savefig(os.path.join(FIG_DIR,"regularisation_example.pdf"))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

np.random.seed(0)

# Generate correlated features
n = 100

x1 = np.random.normal(size=n)
x2 = x1 + np.random.normal(scale=0.1, size=n)  # highly correlated

X = np.column_stack([x1, x2])

# True relationship
y = 3*x1 + np.random.normal(scale=0.5, size=n)

# OLS model
ols = LinearRegression()
ols.fit(X, y)

# Ridge models with different regularisation strengths
alphas = np.logspace(-3, 3, 100)
coefs = []

for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(X, y)
    coefs.append(ridge.coef_)

coefs = np.array(coefs)

# Plot coefficient shrinkage
plt.figure(figsize=(8,5))

plt.plot(alphas, coefs[:,0], label="Coefficient for x1")
plt.plot(alphas, coefs[:,1], label="Coefficient for x2")

plt.xscale("log")

plt.xlabel("Regularisation strength (alpha)")
plt.ylabel("Coefficient value")
plt.title("Ridge Regression Shrinks Coefficients")

plt.legend()
plt.tight_layout()

plt.savefig(os.path.join(FIG_DIR,"ridge_coefficient_shrinkage.pdf"))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso, ElasticNet
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

np.random.seed(0)

# generate correlated features
n = 120
x1 = np.random.normal(size=n)
x2 = x1 + np.random.normal(scale=0.15, size=n)
x3 = x1 + np.random.normal(scale=0.15, size=n)
x4 = np.random.normal(size=n)
x5 = np.random.normal(size=n)

X = np.column_stack([x1, x2, x3, x4, x5])

# true signal depends mainly on first correlated group
y = 3*x1 - 2*x2 + np.random.normal(scale=0.5, size=n)

# fit models
ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=0.1)
enet = ElasticNet(alpha=0.1, l1_ratio=0.5)

ridge.fit(X, y)
lasso.fit(X, y)
enet.fit(X, y)

# coefficient comparison
features = np.arange(X.shape[1])
width = 0.25

plt.figure(figsize=(9,5))
plt.bar(features - width, ridge.coef_, width=width, label="Ridge")
plt.bar(features, lasso.coef_, width=width, label="Lasso")
plt.bar(features + width, enet.coef_, width=width, label="Elastic Net")
plt.xlabel("Feature index")
plt.ylabel("Coefficient value")
plt.title("Coefficient Comparison: Ridge vs Lasso vs Elastic Net")
plt.xticks(features)
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(FIG_DIR, "elastic_net_comparison.pdf"))
plt.show()
```

## Weigh Initialisation

```python
import numpy as np
import matplotlib.pyplot as plt
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

np.random.seed(0)

n_layers = 20
width = 100
n_samples = 200

X = np.random.randn(n_samples, width)

scales = [0.1, 1.0, 2.0]
results = []

for scale in scales:
    activations = X.copy()
    stds = []

    for _ in range(n_layers):
        W = np.random.randn(width, width) * scale / np.sqrt(width)
        activations = np.tanh(activations @ W)
        stds.append(np.std(activations))

    results.append(stds)

layers = np.arange(1, n_layers + 1)

plt.figure(figsize=(8,5))

for scale, stds in zip(scales, results):
    plt.plot(layers, stds, marker='o', label=f"scale = {scale}")

plt.xlabel("Layer")
plt.ylabel("Activation standard deviation")
plt.title("Effect of Weight Initialisation Scale")
plt.legend()
plt.tight_layout()

plt.savefig(os.path.join(FIG_DIR, "weight_initialisation_scale.pdf"))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import os

FIG_DIR = "/Figures/"
os.makedirs(FIG_DIR, exist_ok=True)

np.random.seed(0)

layers = 30
width = 200
samples = 500

X = np.random.randn(samples, width)

def run_network(init_type):

    activations = X.copy()
    stds = []

    for _ in range(layers):

        if init_type == "xavier":
            W = np.random.randn(width, width) * np.sqrt(1/width)

        elif init_type == "he":
            W = np.random.randn(width, width) * np.sqrt(2/width)

        activations = np.maximum(0, activations @ W)  # ReLU

        stds.append(np.std(activations))

    return stds


xavier_stds = run_network("xavier")
he_stds = run_network("he")

layer_idx = np.arange(1, layers+1)

plt.figure(figsize=(8,5))

plt.plot(layer_idx, xavier_stds, label="Xavier initialisation")
plt.plot(layer_idx, he_stds, label="He initialisation")

plt.xlabel("Layer")
plt.ylabel("Activation standard deviation")
plt.title("Xavier vs He Initialisation")

plt.legend()
plt.tight_layout()

plt.savefig(os.path.join(FIG_DIR,"xavier_vs_he_initialisation.pdf"))

plt.show()
```

## Early Stopping

```python
import numpy as np
import matplotlib.pyplot as plt
import os

# Simulated training history over 100 epochs
epochs = np.arange(1, 101)

# Training loss decreases steadily
train_loss = np.exp(-epochs / 35) + 0.02 * np.random.randn(len(epochs))

# Validation loss improves at first, then worsens after about epoch 30
val_loss = np.exp(-epochs / 10) + 0.002 * (epochs - 40)**2 / 100 + 0.02 * np.random.randn(len(epochs))

# Early stopping point
best_epoch = np.argmin(val_loss)

# Plot
plt.plot(epochs, train_loss, label="Training loss")
plt.plot(epochs, val_loss, label="Validation loss")
plt.axvline(best_epoch, linestyle="--", label="Early stopping point")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Early Stopping Example")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated training history over 100 epochs
epochs = np.arange(1, 101)

# Training loss decreases steadily
train_loss = 1.2 * np.exp(-epochs / 35) + 0.05

# Validation loss improves at first, then worsens after about epoch 30
val_loss = 1.0 * np.exp(-epochs / 20) + 0.15 + 0.0025 * np.maximum(epochs - 30, 0)

# Early stopping point
best_epoch = np.argmin(val_loss) + 1
best_val_loss = val_loss[best_epoch - 1]

print(f"Best epoch for early stopping: {best_epoch}")
print(f"Validation loss at best epoch: {best_val_loss:.4f}")

# Plot
plt.figure(figsize=(7, 5))
plt.plot(epochs, train_loss, label="Training Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.axvline(best_epoch, linestyle="--", label=f"Early stopping at epoch {best_epoch}")
plt.scatter(best_epoch, best_val_loss, zorder=3)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Early Stopping Based on Validation Loss")
plt.legend()
plt.tight_layout()
plt.show()
```
